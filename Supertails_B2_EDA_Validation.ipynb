```
import pandas as pd
import matplotlib.pyplot as plt
```

### Load cleaned data
**Why:** EDA must be performed only on cleaned, trusted data.

```
orders = pd.read_csv('orders.csv')
orders['order_date'] = pd.to_datetime(orders['order_date'], errors='coerce')

orders.head()
```

### Order value distribution
**Why:** Understand customer spending behavior and detect outliers.

```
orders['order_value'].describe()
```

```
plt.figure()
plt.hist(orders['order_value'], bins=30)
plt.title('Order Value Distribution')
plt.show()
```

### Revenue by product category
**Why:** Identify top-performing categories driving GMV.

```
category_revenue = orders.groupby('product_category')['order_value'].sum().sort_values(ascending=False)
category_revenue.head()
```

```
plt.figure()
category_revenue.plot(kind='bar')
plt.title('Revenue by Product Category')
plt.show()
```
